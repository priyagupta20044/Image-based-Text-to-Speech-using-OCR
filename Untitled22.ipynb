```python
from PIL import Image, ImageEnhance

# Load the image
im = Image.open("mqdefault.jpg")

# Enhance sharpness
enhancer = ImageEnhance.Sharpness(im)
im = enhancer.enhance(2.0)  # Adjust factor for desired sharpness

# Enhance contrast
enhancer = ImageEnhance.Contrast(im)
im = enhancer.enhance(1.5)   # Adjust factor for desired contrast

# Save the enhanced image
im.save("enhanced_image.jpg")
```

```python
from googletrans import Translator
import easyocr
from gtts import gTTS
from IPython.display import Audio
```

```python
reader = easyocr.Reader(['en'])
translator = Translator()
```

```python
import PIL
from PIL import ImageDraw
im = PIL.Image.open('mqdefault.jpg')
```

```python
im
```

```python
bounds = reader.readtext('mqdefault.jpg',add_margin=0.55,width_ths=0.7,link_threshold=0.8, decoder='beamsearch',blocklist='=-')
bounds
```

```python
def draw_boxes(images,bounds,color='yellow',width=2):
  draw = ImageDraw.Draw(images)
  for bound in bounds:
    p0,p1,p2,p3 = bound[0]
    draw.line([*p0,*p1,*p2,*p3,*p0] , fill=color,width=width)
  return images

draw_boxes(im,bounds)
```

```python
text_list = reader.readtext('mqdefault.jpg', add_margin=0.55, width_ths=0.7, link_threshold=0.9, decoder='beamsearch', blocklist='=-~/_!|')
text_list
```

```python
text_comb = ' '.join([text[1] for text in text_list])
text_comb
```

```python
translator = Translator()
```

```python
text_hin = translator.translate(text_comb,dest='hi')
text_hin.text
```

```python
hi_tts=gTTS(text_hin.text,lang='hi')
hi_tts.save('hi_tts.mp3')
```

```python
Audio('hi_tts.mp3',autoplay=True)
```

```python
text_hin = translator.translate(text_comb,dest='ja')
text_hin.text
```

```python
ja_tts=gTTS(text_hin.text,lang='ja')
ja_tts.save('ja_tts.mp3')
```
